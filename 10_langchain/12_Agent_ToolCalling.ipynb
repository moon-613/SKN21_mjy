````python
tavily_search = TavilySearch()
model = ChatOpenAI(model="gpt-5-mini")
tool_model = model.bind_tools(tools=[tavily_search])   # modelì— toolì„ binding

system_message = """# Instruction
ë‹¹ì‹ ì€ ë‹¤ì–‘í•œ ì§ˆë¬¸ì— ë‹µì„ í•˜ëŠ” Assistantì…ë‹ˆë‹¤. 
ì‚¬ìš©ì ì§ˆë¬¸ì— ëŒ€í•´ ìµœì‹  ì •ë³´ë¥¼ ë°”íƒ•ìœ¼ë¡œ ë‹µë³€í•´ì£¼ì„¸ìš”.
ì˜¤ëŠ˜ ë‚ ì§œëŠ” {today}ì…ë‹ˆë‹¤.
í•„ìš”í•˜ë‹¤ë©´ toolì„ ì´ìš©í•´ ë‹µë³€ì— í•„ìš”í•œ ì •ë³´ë¥¼ ìˆ˜ì§‘í•©ë‹ˆë‹¤.
ë‹¹ì‹ ì´ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” TOOLì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
- tavily_search: ì›¹ ê²€ìƒ‰ íˆ´ """
today = date.today()
prompt = ChatPromptTemplate(
    [
        ("system", system_message),
        ("human", "{query}"),
        MessagesPlaceholder(variable_name="tool_messages", optional=True)
    ], 
    partial_variables={"today":today}
)

tool_model_chain = prompt | tool_model 
````

````python
@chain
def websearch_agent(query:str) -> tuple[str, list]:
    """
    Agent ì „ì²´ íë¦„ì„ ì²˜ë¦¬í•˜ëŠ” Runnable
    ì‚¬ìš©ì ì§ˆë¬¸(query)ë¥¼ ë°›ì•„ì„œ tool_modelê³¼ toolì„ ì´ìš©í•´ì„œ ìµœì¢… ì‘ë‹µì„ ë°˜í™˜í•œë‹¤.
    ReAct êµ¬ì¡°ë¡œ êµ¬ì„±. (ReAct Loop)
    """
    ai_message = tool_model_chain.invoke({"query":query}) # ë°˜í™˜: 1. ì‹¤ì œ ì‘ë‹µ ë˜ëŠ” 2. tool calls ì •ë³´
    messages = []  # tool call ë©”ì„¸ì§€, ToolMessageë“¤ì„ ì €ì¥ ë¦¬ìŠ¤íŠ¸ (prompt tool_messagesì— ë„£ì„ ê²ƒ)
    while ai_message.tool_calls:  # ë¹ˆ ë¦¬ìŠ¤íŠ¸ê°€ ì•„ë‹ˆë¼ë©´
        # Tool í˜¸ì¶œ 
        tool_messages = tavily_search.batch(ai_message.tool_calls) # [TM, TM, ...]
        messages.extend([ai_message, *tool_messages])
        # Tool í˜¸ì¶œ ê²°ê³¼ë¥¼ ë„£ì–´ì„œ tool_modelì— ë‹¤ì‹œ ìš”ì²­
        ai_messages = tool_model_chain.invoke({"query":query, "tool_messages":messages})

    messages.append(ai_message)  # ë§ˆì§€ë§‰ ì‘ë‹µ ê²°ê³¼ ì¶”ê°€
    return ai_message.content  # AIMessages

# agent ë¡œì§ -> LangGraph
````

## @tool ì‚¬ìš©
- í•¨ìˆ˜ë¡œ êµ¬í˜„í•˜ê³  `@tool` ë°ì½”ë ˆì´í„°ë¥¼ ì‚¬ìš©í•´ tool(StructuredTool)ë¡œ ì •ì˜í•œë‹¤.
    - `langchain_core.tools` ëª¨ë“ˆì— ìˆë‹¤.
- tool name
    - í•¨ìˆ˜ì˜ ì´ë¦„ì´ toolì˜ ì´ë¦„ì´ ëœë‹¤.
- parameters
    - í•¨ìˆ˜ì˜ íŒŒë¼ë¯¸í„°ê°€ toolì˜ íŒŒë¼ë¯¸í„°ê°€ ëœë‹¤.
    - **type hint**ë¥¼ ì´ìš©í•´ íƒ€ì…ì„ ì§€ì •í•œë‹¤.  
- description
    - doctringì´ descriptionì´ ëœë‹¤.
    - RunnableBindingì´ toolì„ ì˜ ì°¾ì„ ìˆ˜ ìˆë„ë¡ í•˜ë ¤ë©´ **toolì˜ ê¸°ëŠ¥ì„ ìµœëŒ€í•œ êµ¬ì²´ì **ìœ¼ë¡œ ì‘ì„±í•œë‹¤.
- **@toolì´ ì ìš©ëœ í•¨ìˆ˜(StructuredTool)ì´ tool**ì´ë¯€ë¡œ modelì— binding í•œë‹¤.

````python
from langchain_core.tools import tool

@tool
def plus(num1:int|float, num2:int|float):
    """ë‘ ê°œ ìˆ˜ë¥¼ ë°›ì•„ì„œ ë§ì…ˆí•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” íˆ´"""  # docstring -> tool ì„¤ëª… (description)
    return num1 + num2
@tool
def minus(num1:int|float, num2:int|float):
    """ë‘ ê°œ ìˆ˜ë¥¼ ë°›ì•„ì„œ ëº„ì…ˆí•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” íˆ´"""  
    return num1 - num2
@tool
def multiply(num1:int|float, num2:int|float):
    """ë‘ ê°œ ìˆ˜ë¥¼ ë°›ì•„ì„œ ê³±ì…ˆí•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” íˆ´"""  
    return num1 * num2

# íŒŒë¼ë¯¸í„°ì— ëŒ€í•œ ìì„¸í•œ ì„¤ëª…ì´ ë“¤ì–´ê°€ì•¼ í•˜ëŠ” ê²½ìš° -> Pydantic
from pydantic import BaseModel, Field
class DivideParameterSchema(BaseModel):
    num1: int|float = Field(..., description="ë‚˜ëˆŒ ë•Œ ì‚¬ìš©í•  ì²« ë²ˆì§¸ ê°’")
    num2: int|float = Field(..., description="ë‚˜ëˆŒ ë•Œ ì‚¬ìš©í•  ë‘ ë²ˆì§¸ ê°’")

@tool("divide_tool", args_schema=DivideParameterSchema)  # í•¨ìˆ˜ëª…ê³¼ ___ë¥¼ ë”°ë¡œ ì¤„ ìˆ˜ ìˆë‹¤. 
def divide(num1:int|float, num2:int|float):
    """ë‘ ê°œ ìˆ˜ë¥¼ ë°›ì•„ì„œ ë‚˜ëˆ—ì…ˆí•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” íˆ´"""  
    return num1 / num2
````

## Runnableì„ toolë¡œ ì •ì˜
- `Runnableê°ì²´.as_tool()`
    - name, description, args_schema íŒŒë¼ë¯¸í„°ë¥¼ ì´ìš©í•´ toolì˜ ì´ë¦„, ì„¤ëª…, ìŠ¤í‚¤ë§ˆë¥¼ ì„¤ì •í•œë‹¤.

````python
from langchain_community.document_loaders import WikipediaLoader  # ìœ„í‚¤ë°±ê³¼ ì‚¬ì „ ë‚´ìš©ì„ í¬ë¡¤ë§í•´ì„œ ë¬¸ì„œë¡œ ë°˜í™˜
from pydantic import BaseModel, Field
from langchain_core.runnables import chain
import json

@chain
def wikipedia_search(inputs:dict) -> str:
    """ì‚¬ìš©ì queryë¥¼ ìœ„í‚¤ë°±ê³¼ì‚¬ì „ì—ì„œ ê²€ìƒ‰í•˜ê³  ê²€ìƒ‰ ê²°ê³¼ ì¤‘ kê°œ ë¬¸ì„œë¥¼ ë°˜í™˜í•œë‹¤. 
    Args:
        inputs(dict): query - ê²€ìƒ‰í•  í‚¤ì›Œë“œ, max_results(int): 'ë°˜í™˜í•  ê²€ìƒ‰ ê²°ê³¼ ìˆ˜'ë¥¼ ë°›ëŠ”ë‹¤.
    Returns:
        str - ê²€ìƒ‰ ê²°ê³¼ë¥¼ jsonìœ¼ë¡œ ë°˜í™˜. 
              í˜•ì‹: {"result":[{ê²€ìƒ‰ ë¬¸ì„œ1}, {ê²€ìƒ‰ ë¬¸ì„œ2}, ...]}. í˜•ì‹: {"result":[ê²€ìƒ‰ ë¬¸ì„œ1, ê²€ìƒ‰ ë¬¸ì„œ2, ...]}. ê²€ìƒ‰ ë¬¸ì„œ (dict) - content, url, titleë¡œ êµ¬ì„±
              ê²€ìƒ‰ ê²°ê³¼ê°€ ì—†ì„ ê²½ìš° : {"result":"ê²°ê³¼ ì—†ìŒ"} ë°˜í™˜.
    """

    query = inputs["query"]
    max_results = inputs.get("max_results", 5)
    loader = WikipediaLoader(query=query, load_max_docs=max_results, lang="ko")
    docs = loader.load()

    result_list = []
    for doc in docs:
        result_list.append({"content":doc.page_content, 
                            "title":doc.metadata.get('title'),
                            "url":doc.metadata.get("source")
                            })
        
    if result_list:
        result = {"result": result_list}
    else: 
        result = {"result": "ê²€ìƒ‰ ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤."}

    # JSON  ë¬¸ìì—´ë¡œ ë°˜í™˜í•´ì„œ ë³€í™˜.   # json.load(): json ë¬¸ìì—´ -> python ìë£Œ êµ¬ì¡°, json.dump(): python ìë£Œ êµ¬ì¡° -> json
    return json.dumps(result, ensure_ascii=False)   # ensure_ascii=False : í•œê¸€ ë¬¸ìê°€ ë‚˜ì˜¤ë„ë¡ ì²˜ë¦¬. True: í•œê¸€ì´ ìœ ë‹ˆì½”ë“œ escapeë¬¸ìë¡œ ë‚˜ì˜´ (\uxxxx) 
````

## Vector Store(Vector ì €ì¥ì†Œ) tool

### text loading -> Document ìƒì„±
- ë ˆìŠ¤í† ë‘ ë©”ë‰´ë¥¼ vector storeì— ì €ì¥í•œë‹¤.
1. ë©”ë‰´ text ë¥¼ ë¡œë”©í•œë‹¤.
2. ê° ë©”ë‰´ì˜ ë‚´ìš©(ìŒì‹ì´ë¦„, ë©”ë‰´ì„¤ëª…, íŒŒì¼ëª…)ì„ ë„£ì–´ Documentë¥¼ ìƒì„±í•œë‹¤.

````python
# ë¬¸ì„œ Load
import os
# ë©”ë‰´ì½ì–´ ì˜¤ê¸°
menu_file_path = "data/restaurant_menu.txt"
with open(menu_file_path, "r", encoding="utf-8") as f:
    menu = f.read()
````

````python
# Split - ë©”ë‰´ë³„ë¡œ ë¶„ë¦¬í•˜ê¸° ìœ„í•´ ì§ì ‘ ì²˜ë¦¬
from langchain_core.documents import Document
import re
from pprint import pprint

def create_documents(menu_all: str, file_name) -> list[Document]:
    """ë©”ë‰´ë³„ë¡œ ë¶„ë¦¬ í•œë’¤ Documentë¡œ ë³€í™˜. ë³€í™˜ëœ Documentë“¤ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
    metadata: ë©”ë‰´ì´ë¦„, ì£¼ìš” ì¬ë£Œ, ê°€ê²©, source(íŒŒì¼ì´ë¦„)
    """
    menu_list = menu_all.split("\n\n")
    menu_documents = []
    for menu_item in menu_list:

        # ë©”ë‰´ ì´ë¦„ ì¶”ì¶œ
        menu_name_pattern = r"(^\d+.\s)([ê°€-í£a-zA-Z\d ]+)"
        menu_name_result = re.search(menu_name_pattern, menu_item)
        menu_name = menu_name_result.group(2).strip()  # ë©”ë‰´ ì´ë¦„ ì¶”ì¶œ # 1ì€ ë²ˆí˜¸

        # ì£¼ìš” ì¬ë£Œ ì¶”ì¶œ - textë¡œ indexing.
        menu_ingredient_pattern = r"(?<=ì£¼ìš” ì¬ë£Œ: ).+"
        ingredients = re.search(menu_ingredient_pattern, menu_item).group().strip()
    
        # ê°€ê²© ì¶”ì¶œ
        price_pattern = r"(?<=ê°€ê²©: )[\d,]+(?=ì›)"
        price_result = re.search(price_pattern, menu_item)
        price = int(price_result.group().replace(",", ""))

        
        # ë©”ë‰´ ì´ë¦„ì„ ì œì™¸í•œ ë‚˜ë¨¸ì§€ëŠ” ë©”ë‰´ ì„¤ëª…ìœ¼ë¡œ ê°„ì£¼
        menu_doc = Document(
            page_content=menu_item,     # ë©”ë‰´ ì„¤ëª…ì„ page_contentì— ë„£ëŠ”ë‹¤. metadataëŠ” ë‚˜ì¤‘ì— ë”°ë¡œ ë”°ë¡œ payload filtering í•  ìˆ˜ìˆê¸° ë•Œë¬¸ì— 
            metadata={
                "menu_name": menu_name, # ë©”ë‰´ ì´ë¦„
                "price": price,         # ê°€ê²©
                "ingredients": ingredients, # ì¬ë£Œ ë¦¬ìŠ¤íŠ¸
                "source": file_name     # ë©”ë‰´ê°€ ì €ì¥ëœ íŒŒì¼ ì´ë¦„
            }
        )
        menu_documents.append(menu_doc)
    return menu_documents


menu_documents = create_documents(menu, menu_file_path)
print(len(menu_documents))
menu_documents[0]
````

````python
# VectorDB ì €ì¥ ë° VectorStore ìƒì„±
from langchain_qdrant import QdrantVectorStore
from langchain_openai import OpenAIEmbeddings
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams
from dotenv import load_dotenv

load_dotenv()

COLLECTION_NAME = "restaurant_menu"
VECTOR_SIZE = 1536  # OpenAIEmbeddingsì˜ ë²¡í„° í¬ê¸°

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

client = QdrantClient(host="localhost", port=6333)
if client.collection_exists(COLLECTION_NAME):
    client.delete_collection(COLLECTION_NAME)

# Collection ìƒì„±
client.create_collection(
    collection_name=COLLECTION_NAME,
    # ë²¡í„° ì €ì¥ì†Œì— ëŒ€í•œ ì„¤ì •. - vector_configëŠ” defaultì´ë¯€ë¡œ ì´ë¦„ì„ ì•ˆì¤˜ë„ ë¨(default ë¡œ ì¡í˜) ì´ë¦„ì„ ì¤„ ê±°ë©´ {"ì´ë¦„":VectorParams()} ì„¤ì •
    vectors_config=VectorParams(
        size=VECTOR_SIZE, 
        distance=Distance.COSINE
    )
)

vectorstore = QdrantVectorStore(
    client=client, # QdrantClient
    embedding=embeddings, # Embedding Model
    collection_name=COLLECTION_NAME # ì—°ê²°í•  Collection ì§€ì •.
)
ids = vectorstore.add_documents(menu_documents)
print(f"ì €ì¥ëœ ë¬¸ì„œ ê°œìˆ˜: {len(ids)}")
##################################################
# Retriever ìƒì„±
##################################################
menu_retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}  # ê²€ìƒ‰í•  ê²°ê³¼ ê°œìˆ˜
)

from langchain_core.runnables import ConfigurableField
retriever = menu_retriever.configurable_fields(
    search_type=ConfigurableField(
        id="search_type"
    ),
    search_kwargs=ConfigurableField(
        id="search_kwargs"
    ),
)
````

````python
query = "í”„ë‘ìŠ¤ ìŒì‹ì„ ì¶”ì²œí•´ì¤˜."
query = "ìŠ¤í˜ì¸ ìŒì‹ì„ ì¶”ì²œí•´ì¤˜."
result = retriever.invoke(query)

for doc in result:
    print(doc.metadata)
````

````python
from langchain_core.tools import tool
# Retrieverë¥¼ toolë¡œ ì‚¬ìš©.
# queryë¥¼ ì…ë ¥ ë°›ì•„ì„œ retrieverë¡œ ê²€ìƒ‰í•´ì„œ ê·¸ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” toolì„ ì§ì ‘ êµ¬í˜„.

@tool
def search_menu(query:str) -> str:
    """VectorStore(Vector Database)ì— ì €ì¥ëœ restaurant menuë¥¼ ê²€ìƒ‰í•˜ëŠ” tool.
    ì´ ë„êµ¬ëŠ” queryì— ë§ëŠ” ìŒì‹ ë©”ë‰´ë¥¼ ê²€ìƒ‰í•  ë•Œ ì‚¬ìš©í•œë‹¤.

    Args:
        query(str): vector dbì˜ restaurant ë©”ë‰´ì—ì„œ ê²€ìƒ‰í•  ìŒì‹ì— ëŒ€í•œ query.
    Return:
        str: ê²€ìƒ‰ ë©”ë‰´ë“¤ì— ëŒ€í•œ ì •ë³´ë¥¼ JSON í˜•ì‹ìœ¼ë¡œ ë°˜í™˜í•œë‹¤.
    """
    result_list = []
    docs = retriever.invoke(query)
    for doc in docs:
        result_list.append({"content":doc.page_content,
                            "title":doc.metadata['menu_name'],
                            "url": doc.metadata['source']})
    if result_list:
        result = {"result":result_list}
    else:
        result = {"result": "ê²€ìƒ‰ëœ ë©”ë‰´ê°€ ì—†ìŠµë‹ˆë‹¤."}
    
    return json.dumps(result, ensure_ascii=False)
````
